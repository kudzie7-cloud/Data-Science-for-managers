```python
#Importing required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px


```

```python
#Load/Import dataset into Jupyter Notebook
superstore = pd.read_excel("Sample - Superstore.xls")

#Show the first 5 rows of the imported data
superstore.head()
```

```python
#Display information about dataset and check for missing values
superstore.info()
```

```python
#Display summary statistics for numeric columns
superstore.describe()
```

```python
# group by category and calculate mean of values
grouped_data = superstore.groupby('State')['Profit'].mean()

grouped_data.head()
```

```python
grouped_data.describe()
```

```python
grouped_data.info()
```

```python
# create a sample dataframe
data = {'category': superstore['State'],
        'values': superstore['Sales']}
df = pd.DataFrame(data)

# group by category and calculate mean of values
grouped_df = df.groupby('category')['values'].mean().reset_index()

# create a pie chart
plt.figure(figsize=(18, 14))
plt.pie(grouped_df['values'], labels=grouped_df['category'], autopct='%1.1f%%')
plt.title('Mean Values by Category')
plt.show()
```

```python
#Using Plotly to create an interactive pie chart
fig = px.pie(data, values='values', names='category')

# Add a title to the pie chart
fig.update_layout(title='Pie Chart for Sales By State')
fig.update_layout(width=1000, height=800)
# Show the pie chart
fig.show()
```

```python
# Create a count plot by category
fig = px.bar(data, x='category',
             y='values',
             title='Count Plot by Category')
fig.update_layout(width=1000, height=600)
fig.update_traces(marker_color='green')

# Show the plot
fig.show()
```

```python
# create a sample dataframe
data2 = {'category': superstore['Sub-Category'],
        'values': superstore['Sales']}
df2 = pd.DataFrame(data2)

# group by category and calculate mean of values
grouped_df2 = df2.groupby('category')['values'].mean().reset_index()

# create a pie chart
plt.figure(figsize=(18, 14))
plt.pie(grouped_df2['values'], labels=grouped_df2['category'], autopct='%1.1f%%')
plt.title('Mean Values by Category')
plt.show()
```

```python
# Create a count plot by category
fig = px.bar(data2, x='category', y='values', title='Count Plot by Category')
fig.update_layout(width=1000, height=600)

# Show the plot
fig.show()
```

```python
plt.figure(figsize=(12,30))
sns.barplot(x= 'Sales',
           y= 'State',
           hue= 'Category',
           data= superstore)
plt.title('Barplot showing Sales by state')
plt.xlabel('Sales')
plt.ylabel('State')
plt.show()
```
